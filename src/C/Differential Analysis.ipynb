```python
import numpy as np
```

```python
def compute(alpha_i, alpha_i1, x_i, x_i1):
    return ((alpha_i*x_i1) ^ alpha_i1*x_i ^ alpha_i*(1^alpha_i1))
```

```python
def List_Entries(p):
    output = []
    for i in range(2**p):
        val = [int(j) for j in np.binary_repr(i, p)]
        output.append(val)
    return output
```

```python
def compute_case(alpha, p):
    output = 0
    tmp = []
    for i in range(2**p):
        entry = [int(j) for j in np.binary_repr(i, p)]
        out = [compute(alpha[i], alpha[(i+1)%p], entry[i], entry[(i+1)%p]) for i in range(p)]
        if out not in tmp:
            tmp.append(out)
            output += 1
    return output, tmp
```

```python
def possibility(p):
    for i in range(2**p):
        case = [int(j) for j in np.binary_repr(i, p)]
        c,_ = compute_case(case, p)
        print("case {} ==> {}".format(case, c))
```
